```python
df = fetch_financial_data('UBER')
df.info()
```

```python
df.head()
```

```python
df['Average']=(df.Open +df.Close)
df.head(3)
```

```python
df = df.sort_index()
df.head(3)
```

```python
df[['Close']].shift(3)
df.head(3)
```

```python
df['Close_shift'] = df.Close.shift(1)
df.head(3)
```

```python
df['Daily_change'] = (df.Close / df.Close_shift - 1)*100
df.head(3)
```

```python
df.Daily_change.min()
```
